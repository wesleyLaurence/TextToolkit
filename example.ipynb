# TextToolkit
TextToolkit is a library for mining unstructured text. The Analyzer class provides methods for text data processing, cleaning, word frequency analysis and collocation. In this notebook, we will be analyzing text from a research paper I wrote, titled _"Music Cognition: Music's Effect on the Brain from Infancy to Adolesence"_.
    - Analysis by Wesley Laurence

```python
# import library
import text_toolkit.analyze as a
```

```python
# load text into Analyzer
music_cognition = a.Analyzer("music-cognition-data.txt")
```

### Word Frequency

```python
music_cognition.visual_word_count()
```

```python
music_cognition.word_count().head(20)
```

### Collocation

```python
# visualize bigram
music_cognition.visual_collocate(2)
```

```python
# bigram
music_cognition.collocate(2).head(20)
```

```python
# visualize trigram
music_cognition.visual_collocate(3)
```

```python
# trigram
music_cognition.collocate(3).head(20)
```
